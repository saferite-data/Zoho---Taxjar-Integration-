## Zoho books: documentos API

La API de Zoho Books le permite realizar todas las operaciones que realiza con nuestro cliente web.
La API de Zoho Books se creó con principios REST que garantizan URL predecibles que facilitan la creación de aplicaciones. Esta API sigue las reglas de HTTP, lo que permite que se pueda utilizar una amplia gama de clientes HTTP para interactuar con la API.
Cada recurso se expone como una URL. La URL de cada recurso se puede obtener accediendo al Punto final raíz de la API.

Límite de llamadas de la API
Las llamadas a la API están limitadas para brindar una mejor calidad de servicio y disponibilidad a todos los usuarios. Los límites en el total de solicitudes por día se enumeran a continuación para cada plan:

Plan gratuito - 1000 solicitudes de API/día
Plan Estándar- 2000 solicitudes/día
Plan Profesional - 5000 solicitudes/día
Plan Premium- 10000 solicitudes/día
Plan Elite- 10000 solicitudes/día
Ultimate Plan- 10000 solicitudes/día

## Identificación de la organización (Organization ID)


En Zoho Books, su empresa se denomina organización. 

Si tiene varias empresas, simplemente configure cada una de ellas como una organización individual. Cada organización es una organización de Zoho Books independiente con su propio ID de organización(Organization ID), moneda base, zona horaria, idioma, contactos, informes, etc.

El parámetro <b>organization_id</b> junto con el ID de la organización debe enviarse con cada solicitud de API para identificar la organización.

## OAuth

Las API REST de Zoho utilizan el protocolo OAuth 2.0 para autorizar y autenticar llamadas. Brinda acceso seguro para proteger los recursos, lo que reduce la molestia de solicitar un nombre de usuario y una contraseña cada vez que un usuario inicia sesión. Siga los pasos que se enumeran aquí para acceder a las API de Zoho mediante OAuth 2.0

Nota: Las URL de la API en esta sección deben modificarse según su dominio.

{Centro de datos:Estados Unidos,	Dominio:.com,	URI de la API base: https://accounts.zoho.com/}

### Paso 1: Registro De Nuevo Cliente

Primero deberá registrar su aplicación con la consola de desarrollador de Zoho para obtener su Client IDy Client Secret.

Para registrar su aplicación, vaya a https://accounts.zoho.com/developerconsole y haga clic en Add Client ID. Proporcione los detalles requeridos para registrar su aplicación. Si se registra correctamente, se le proporcionará un conjunto de credenciales de OAuth 2.0, como un Client ID y, Client Secret que Zoho y su aplicación conocen. No comparta estas credenciales en ningún lado.

### Paso 2: Generación De Token De Concesión (Grant Token)
Redirigir a la siguiente URL de autorización con los parámetros dados

https://accounts.zoho.com/oauth/v2/auth?

example url api paramts request:

https://accounts.zoho.com/oauth/v2/auth?scope=ZohoBooks.invoices.CREATE,ZohoBooks.invoices.READ,ZohoBooks.invoices.UPDATE,ZohoBooks.invoices.DELETE&client_id=1000.0SRSZSY37WMZ69405H3TMYI2239V&state=testing&response_type=code&redirect_uri=http://www.zoho.com/books&access_type=offline

<b>scope *</b> = SCOPE para el cual se generará el token. Se pueden dar múltiples scope deben estar separados por comas. Ex : ZohoBooks.fullaccess.all

<b>client_id *</b> = Client ID obtenido durante el registro del cliente


<b>state</b> = un opaque string que es de ida y vuelta en el protocolo; es decir, cualquier valor que se le dé a esto se le retornara a usted.

<b>response_type</b> = code

<b>redirect_uri</b> =  URI de redirección proporcionados en el paso anterior. Este parámetro debe ser la misma URL de redireccionamiento mencionada al registrar al Cliente

<b>access_type</b> = Los valores permitidos son offliney online. access_type online le da a su aplicación solo el access_token que es válido por una hora. access_type offlinele dará a la aplicación un access_token así como un refresh_token. Por defecto se toma como online

<b>prompt</b> = Solicita el consentimiento del usuario cada vez que su aplicación intenta acceder a las credenciales del          usuario. Ex:Consent

<b>Note: Fields with * are mandatory</b>

En esta solicitud(request), se le mostrará una "página de consentimiento del usuario".

Al hacer clic en "Aceptar", Zoho redirigirá al redirect_uri dado con el "code" y "stateparam". Este valor de código es obligatorio para obtener el token de acceso en el siguiente paso y este código es válido durante 60 segundos.


## Paso 3: Generar Token De Acceso y Actualización

Después de realizar code el paso anterior, realice una solicitud POST para la siguiente URL con los parámetros dados, para generar el archivo <b>access_token</b>

https://accounts.zoho.com/oauth/v2/token? + parametros :

ejemplo url con parametros:

https://accounts.zoho.com/oauth/v2/token?code=1000.dd7e47321d48b8a7e312e3d6eb1a9bb8.b6c07ac766ec11da98bf6a261e24dca4&client_id=1000.0SRSZSY37WMZ69405H3TMYI2239V&client_secret=fb0196010f2b70df8db2a173ca2cf59388798abf&redirect_uri=http://www.zoho.com/books&grant_type=authorization_code

parametros = code* , client_id*, scope, state (los mismos del paso anterior)

paramtros nuevos: 

<b>client_secret*</b> = Secreto del cliente obtenido durante el registro del cliente.
<b>redirect_uri*</b> = Este parámetro debe ser la misma URL de redireccionamiento mencionada al agregar el Cliente.
<b>grant_type*</b> = authorization_code

En la respuesta, obtendrá tanto <b>access_token</b> como <b>refresh_token</b>.

1. El access_token testamento expirará después de un período determinado (como se indica en expires_inel parámetro de la respuesta).

2. El refresh_tokenes permanente y se usará para regenerar nuevo access_token, si el token de acceso actual está vencido.

Nota: Cada vez que se acepta una página de nuevo consentimiento, se genera un nuevo token de actualización. El límite máximo es de 20 tokens de actualización por usuario. Si se cruza este límite, el primer token de actualización se elimina automáticamente para acomodar el último. Esto se hace independientemente de si el primer token de actualización está en uso o no.


## Paso 4: Generar Token De Acceso A Partir De Token De Actualización


Los tokens de acceso tienen una validez limitada. En la mayoría de los casos generales, los tokens de acceso caducan en una hora. Hasta entonces, el token de acceso tiene un uso ilimitado. Una vez que caduque, su aplicación deberá usar el token de actualización para solicitar un nuevo token de acceso. Redirigir a la siguiente URL POST con los parámetros proporcionados para obtener un nuevo token de acceso.

https://accounts.zoho.com/oauth/v2/token? + PARAMETERS

Ejemplo

https://accounts.zoho.com/oauth/v2/token?refresh_token=1000.8ecd474019e31d52d2f94aad6c5cb7.4638677ebc14f2f2ee0b6dfb6cebdc&client_id=1000.0SRSZSY37WMZ69405H3TMYI2239V&client_secret=fb0196010f2b70df8db2a173ca2cf59388798abf&redirect_uri=http://www.zoho.com/books&grant_type=refresh_token

PARAMETERS: client_id, client_secret, redirect_uri

refresh_token = REFRESH TOKEN  que se obtiene en el paso anterior

grant_type = refresh_token


## Paso 5: Revocación De Un Token De Actualización

Para revocar un token de actualización, llame a la siguiente URL POST con los parámetros dados

https://accounts.zoho.com/oauth/v2/token/revoke? + PARAMETERS

Ejemplo
https://accounts.zoho.com/oauth/v2/token/revoke?token=1000.8ecd474019e31d522f94aad6c5cb7.4638677ebc14f2f2ee0b6dfb6cebdc

PARAMETERS REFRESH TOKEN que se va a revocar


## Paso 6: Llamar A Una API

El nombre del header debe ser "Authorization"
El valor del header debe ser "Zoho-oauthtoken {access_token}"
