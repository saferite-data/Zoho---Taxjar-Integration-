## ( Taxjar : documentos API )

La API de TaxJar proporciona reglas de impuestos a nivel de producto para un subconjunto de categorías de productos. Estas categorías se utilizarán para productos que están exentos del impuesto sobre las ventas en algunas jurisdicciones o están gravados con tasas reducidas. No necesita pasar un código de impuestos del producto para los cálculos de impuestos sobre las ventas en productos que están totalmente sujetos a impuestos. Simplemente deje ese parámetro fuera.

Cliente oficial de Python para la API de TaxJar . Para obtener la documentación de la API, visite https://developers.taxjar.com/api.

Asegúrese de reemplazar 9e0cd62a22f451701f29c3bde214con su clave API.

TaxJar utiliza claves de API para permitir el acceso a la API. Si es nuevo en TaxJar, deberá registrarse para obtener una cuenta para obtener su clave API. De lo contrario, inicie sesión y vaya a Cuenta > Acceso API para generar un nuevo token de API.

TaxJar espera que la clave de API se incluya en todas las solicitudes de API al servidor mediante un encabezado como el siguiente:

Authorization: Token token="9e0cd62a22f451701f29c3bde214"

o

Authorization: Bearer 9e0cd62a22f451701f29c3bde214
